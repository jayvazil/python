fitering

```python
# prompt: code for data filtering

# Assuming 'df' is your DataFrame (as defined in the previous code)

# Example 2: Filter rows where 'BMI' is between 25 and 30 (inclusive)
filtered_df = df[(df['BMI'] >= 25) & (df['BMI'] <= 30)]
print(filtered_df)

# Example 3: Filter rows where 'Outcome' is 1 (positive for diabetes)
positive_diabetes_df = df[df['Outcome'] == 1]
print(positive_diabetes_df)

# Example 4: Filter out rows where 'BloodPressure' is missing (NaN)
df_no_missing_bp = df.dropna(subset=['BloodPressure'])
print(df_no_missing_bp)

```

applying funtion

```python
def age_cartegory(age):
  if age<30:
    return 'Young'
  else:
      return 'Old'
df['Age_Category']=df['Age'].apply(age_cartegory)
```

matplotlib

```python
plt.hist(df['Age'], bins=20, color='lime', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Sugar'],df['Insulin'], alpha=0.5, c='purple')
plt.xlabel('Sugar')
plt.ylabel('Insulin')
plt.title('Sugar vs Insulin')
plt.show()
```

```python
outcome_counts = df['Outcome'].value_counts()
labels=['No Diabetes','Diabetes']
colors=['blue','orange']
plt.figure(figsize=(8,8))
plt.pie(outcome_counts, labels=labels, colors=colors, autopct='%1.1f%%',startangle=140,wedgeprops={'edgecolor':'black'})
plt.title('Diabetes Outcome Proportion')
plt.show()
```
