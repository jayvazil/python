```python
bds=[45,25,15]
gender=['male','female','intersex']
col=['orange','lime','blue']
expl=[0,0,0.1]
```

```python
myd={"male":45,"female":25}
import matplotlib.pyplot as plt
plt.title("An Example")
plt.pie(x=bds,labels=gender,colors=col,explode=expl)
plt.show()
```

```python
plt.bar(x=gender,height=bds,color=col)
plt.show()
```

```python
plt.barh(y=gender,width=bds,color=col)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
loss=[15,26,45,59,55,67,80]
plt.ylabel("Financial Loss")
plt.xlabel("Day")
plt.title("Loss vs Day")
plt.plot(day,loss)
plt.show()
```
