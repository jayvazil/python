```python
gender_data=df['Sex'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(gender_data.values,labels=gender_data.index,autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
residence=df["residence"].value_counts()
residence
```

```python
plt.barh(residence.index,residence.values)
plt.xlabel('Count')
plt.ylabel('Region')
plt.title('Region Distribution')
plt.show()
```

```python
dwh=df[["weight","height"]]
```

```python
plt.scatter(df['height'],df['weight'],marker='o')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.grid(True)
plt.show()
```

```python
cm=dwh.corr()
print(cm)
```

```python
# prompt: hmap

import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
